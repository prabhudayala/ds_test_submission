```python
#stacked bar plots matplotlib: https://matplotlib.org/gallery/lines_bars_and_markers/bar_stacked.html
def stack_plot(data, xtick, col2='Dependent_Variable', col3='total'):
    ind = np.arange(data.shape[0])
    
    plt.figure(figsize=(20,5))
    p1 = plt.bar(ind, data[col3].values)
    p2 = plt.bar(ind, data[col2].values)

    plt.ylabel('Leads')
    plt.title('% of leads')
    plt.xticks(ind, list(data[xtick].values))
    plt.legend((p1[0], p2[0]), ('total', 'dependent variable'))
    plt.show()
```

```python
### Helper function to plot ####
def univariate_barplots(data, col1, col2='Dependent_Variable', top=False):
    #Count number of zeros in dataframe python:https://stackoverflow.com/a/51540521/4084039
    temp = pd.DataFrame(data.groupby(col1)[col2].agg(lambda x: x.eq(1).sum())).reset_index()

    #Pandas dataframe grouby count:https://stackoverflow.com/a/19385591/4084039
    temp['total'] = pd.DataFrame(data.groupby(col1)[col2].agg(total='count')).reset_index()['total']
    temp['Avg'] = pd.DataFrame(data.groupby(col1)[col2].agg(Avg='mean')).reset_index()['Avg']
    
    temp.sort_values(by=['total'],inplace=True, ascending=False)
    
    if top:
        temp = temp[0:top]
    
    stack_plot(temp, xtick=col1, col2=col2, col3='total')
    print(temp.head(5))
    print("="*50)
```

```python
from scipy import stats
z = np.abs(stats.zscore(X_numerical))
print(X_numerical.shape)
TEMP = X_numerical[(z < 3).all(axis=1)]
print(TEMP.shape)
```

### Kbest feature selection

#### chi2

```python
from sklearn.feature_selection import SelectKBest,chi2,mutual_info_classif
```

```python
def select_features(X_train, y_train, X_test):
    fs = SelectKBest(score_func=chi2, k='all')
    fs.fit(X_train, y_train)
    X_train_fs = fs.transform(X_train)
    X_test_fs = fs.transform(X_test)
    return X_train_fs, X_test_fs, fs
```

```python
data.drop(columns=['is_outlier'],inplace=True)
print(data.shape)
y = data['Dependent_Variable']
X=data.drop(columns=['Dependent_Variable'])
print(X.shape)
```

```python
X_categorical = X[['C1','C2','C3','C4','C5','C6','C7','C8']]
```

```python
#https://machinelearningmastery.com/feature-selection-with-categorical-data/
X_train_fs, X_test_fs, fs = select_features(X_categorical, y, X_categorical)
# what are scores for the features
for i in range(len(fs.scores_)):
    print('Feature %d: %f' % (i, fs.scores_[i]))
# plot the scores
plt.bar([i for i in range(len(fs.scores_))], fs.scores_)
plt.show()
```

#### mutual_info_classif

```python
def select_features(X_train, y_train, X_test):
    fs = SelectKBest(score_func=mutual_info_classif, k='all')
    fs.fit(X_train, y_train)
    X_train_fs = fs.transform(X_train)
    X_test_fs = fs.transform(X_test)
    return X_train_fs, X_test_fs, fs
X_train_fs, X_test_fs, fs = select_features(X_categorical, y, X_categorical)
# what are scores for the features
for i in range(len(fs.scores_)):
    print('Feature %d: %f' % (i, fs.scores_[i]))
# plot the scores
plt.bar([i for i in range(len(fs.scores_))], fs.scores_)
plt.show()
```

**`From above 2 graphs I have decided to remove C3,C6,C8`**
